```python
# Import package 
import swehockey.swehockey_scraper as swe
import numpy as np
import pandas as pd
```

```python
# Read in needed schedule-ids
df_swehockey = pd.read_csv("https://raw.githubusercontent.com/msjoelin/swehockey_scraper/master/data/scheduleid.csv", 
                           error_bad_lines=False,
                          dtype=str)

df_swehockey.head()
```

```python
# get games for some schedule ids
games = swe.getGames(df_swehockey.head())
games.head()
```

```python
# Clean up the output with cleanGames
df_games_clean = swe.cleanGames(games.copy())
df_games_clean.head()
```

```python
# Create dataframe on teamlevel
df_teams = swe.getTeamData(df_games_clean)
df_teams.head()
```

```python
# Get game details
game_events, game_summary = swe.getGameData(games['game_id'].head())
```
